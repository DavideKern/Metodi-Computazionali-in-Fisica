# Trasformazione di una variabile casuale
## Caso di funzioni non invertibili

Sia $x$ una variabile casuale con funzione densità di probabilità $p(x),\; x \in [a,b]$. Sia $\varphi: x \in [a,b] \rightarrow y=\varphi(x) \in \mathbb{R}$ una funzione derivabile a tratti. Vogliamo determinare la funzione densità di probabilità $f(y)$ della variabile casuale $y=\varphi(x)$. $\varphi$ potrebbe non essere invertibile: per ora, supponiamo che $\exists c \in (a,b)$ per cui le restrizioni $\varphi_1,\varphi_2$ di $\varphi$ rispettivamente in $[a,c],[c,b]$ siano invertibili (è il caso ad esempio di una funzione parabolica di vertice $x_V=c$). Si ha quindi che

$\varphi_1 = \varphi|_{[a,c]}: x \in [a,c] \rightarrow y=\varphi(x) \in \varphi \left( [a,c] \right) \\
\varphi_2 = \varphi|_{[c,b]}: x \in [c,b] \rightarrow y=\varphi(x) \in \varphi \left( [c,b] \right)$

$\exists \varphi_1^{-1} \equiv \eta_1 : y \in \varphi \left( [a,c] \right) \rightarrow x=\eta_1(y) \in [a,c] \\
\exists \varphi_2^{-1} \equiv \eta_2 : y \in \varphi \left( [c,b] \right) \rightarrow x=\eta_2(y) \in [c,b]$

Se $\varphi$ è continua in $[a,b]$, $\varphi_1$ e $\varphi_2$ hanno una diversa monotonia (mentre non è difficile trovare esempi di funzioni $\varphi$ discontinue le cui restrizioni invertibili hanno la stessa monotonia. Si pensi al dente di sega): possiamo senza perdere di generalità assumere $\varphi_1$ crescente (quindi lo è $\eta_1$) e $\varphi_2$ decrescente (quindi lo è $\eta_2$). Se $\varphi_{1,2}$ sono derivabili a tratti, lo sono $\eta_{1,2}$ e varrà $\eta_1'(y)<0, \; \eta_2'(y)>0$.

Il dominio della variabile casuale $y=\varphi(x)$ è $\varphi([a,b])$. Distinguiamo tre casi:
1. $y \in \varphi([a,c]) \setminus \varphi([c,b])$
2. $y \in \varphi([a,c]) \cap \varphi([c,b])$
3. $y \in \varphi([c,b] \setminus \varphi([a,c]))$

L'insieme del secondo caso è non vuoto, essendo $\varphi$ non invertibile (una stessa $y$ può essere fissata sia da un punto $x_1 \in [a,c]$ che da uno punto $x_2 \in [c,b]$). L'insieme del primo e/o del terzo caso può invece anche essere vuoto. Si può ad esempio pensare alla seguente funzione:

-

1. In questo caso $\exists!x \in [a,b]: \varphi(x)=y$. In particolare, $x \in [a,c]$ per costruzione. La conservazione della probabilità impone

$P[y \leqslant y' \leqslant y+dy] = P[x \leqslant x' \leqslant x+dx] \;\; \Rightarrow \;\; f(y)dy = p(x)dx$

Ricordando gli insiemi di appartenenza di $x,y$ possiamo riportare la scrittura in funzione di $y$ con la sostituzione $x \equiv \eta_1(y) \;\; \Rightarrow \;\; dx=\eta_1'(y)dy \;$:

$f(y)dy = p(\eta_1(y)) \; \eta_1'(y)dy$

$ \fbox{$ f(y) = p(\eta_1(y)) \; \eta_1'(y) \;\;\;\; y \in \varphi([a,c]) \setminus \varphi([c,b]) $}$

2. In questo caso $\exists x_1 \in [a,c], x_2 \in [c,b]: \varphi(x_1)=\varphi(x_2)=y$. La probabilità si conserva ricordando che $y$ è mappata da due valori diversi delle $x$, quindi a $f(y)$ concorranno due contributi:

$P[y \leqslant y' \leqslant y+dy] = P[x_1 \leqslant x' \leqslant x_1+dx_1]+P[x_2 \leqslant x' \leqslant x_2+dx_2] \;\; \Rightarrow \;\; f(y)dy = p(x_1)dx_1 + p(x_2)dx_2$

In tal caso sostituiamo $x_1 \equiv \eta_1(y),\; x_2 \equiv \eta_2(y) \;\; \Rightarrow \;\; dx_1 = \eta_1'(y)dy; \; dx_2= -\eta_2'(y)dy \;\;$ (essendo $\eta_2'<0$)

$f(y)dy = \left[ p(\eta_1(y)) \; \eta_1'(y) - p(\eta_2(y)) \; \eta_2'(y) \right] dy \;\;$ con $\eta_1'>0, \eta_2'<0$. Possiamo quindi scrivere

$\fbox{$ f(y) = p(\eta_1(y)) \; |\eta_1'(y)| + p(\eta_2(y)) \; |\eta_2'(y)|$} $

3. In analogia al primo caso, $\exists!x \in [a,b]: \varphi(x)=y$, ma stavolta $x \in [c,b]$. Con le stesse argomentazioni di prima, ricordando però che $\eta_2'(y)<0$, si ricava

$f(y)dy = -p(\eta_2(y)) \; \eta_2'(y)dy \\
\fbox{$ f(y) = p(\eta_2(y)) \; |\eta_2'(y)| $}$

In generale, notiamo che il contributo di $p(\eta_1(y))$ figura solo per $y \in \varphi([a,c])$; allo stesso modo, quello di $p(\eta_2(y))$ è presente solo per $y \in \varphi([c,b])$. Possiamo rendere più compatta la notazione estendendo $\eta_1, \eta_2$ alle seguenti funzioni:

$\displaystyle \eta_1(y) \equiv \begin{cases}\varphi_1^{-1}(y): \;\; y \in \varphi([a,c]) \\ 0\;\;\;\;\;\;\;\, : \;\; altrove \end{cases} \; ; \;\; \eta_2(y) \equiv \begin{cases}\varphi_2^{-1}(y): \;\; y \in \varphi([c,b]) \\ 0\;\;\;\;\;\;\;\, : \;\; altrove \end{cases}$

Con questa convenzione otteniamo che $f(y)$ è definita per casi, ma si può comunque esprimere con la seguente scrittura:

$\fbox{$ f(y) = p(\eta_1(y)) \; |\eta_1'(y)| + p(\eta_2(y)) \; |\eta_2'(y)|  \;\;\;\;\; y\in \varphi([a,b])$}$

L'idea si generalizza per funzioni $\varphi$ che oscillano più volte, o comunque per le quali è necessaria una partizione del dominio in più intervalli di invertibilità:

Sia $\varphi: x \in [a,b] \rightarrow y \in \varphi([a,b])$ e sia $a \equiv x_0<x_1<\cdots<x_{n-1}<x_n \equiv b$ una sequenza di punti per cui le restrizioni di $\varphi$ in $[x_0,x_1], \cdots, [x_{n-1},x_n]$ sono invertibili. Definiamo allora

$\displaystyle \begin{cases} \varphi_i \equiv \varphi|_{[x_{i-1},\; x_i]}: x \in [x_{i-1},x_i] \rightarrow \varphi(x) \in \varphi([x_{i-1},x_i]) \\ \eta_i \equiv \varphi_i^{-1}: y \in \varphi([x_{i-1},x_i]) \rightarrow x=\eta_i(y) \in [x_{i-1},x_i] \end{cases} \;\;\;\; \forall i \in \{1, \cdots, n\}$

Anche in questo caso, a seconda di dove si trovi $y$ in $\varphi([a,b])$, nella funzione di probabilità figureranno diversi contributi legati ciascuno a un'opportuna $\eta_i$: in base a quali intervalli $[x_{i-1},x_i]$ includono le $x_i$ mappate da $\varphi$ in $y$, compariranno i rispettivi termini $p(\eta_i(y))\; \eta_i'(y)$ (o $-\eta_i'(y)$, a seconda se $\varphi_i$ è crescente o decrescente in $[x_{i-1},x_i]$). Avendo al solito definito i prolungamenti

$\displaystyle \eta_i(y) \equiv \begin{cases}\varphi_i^{-1}(y): \;\; y \in \varphi([x_{i-1},x_i]) \\ 0 \;\;\;\;\;\;\;\, : \;\; altrove \end{cases} \;\;\;\; \forall i \in \{1, \cdots, n\}$

possiamo finalmente scrivere la funzione densità di probabilità di $y=\varphi(x)$:

$ \displaystyle \fbox{$ f(y) = \sum_{i=1}^n p(\eta_i(y)) \; |\eta_i'(y)| \;\;\;\;\; y \in \varphi([a,b]) $}$

L'espressione di $f$ è quindi generalmente complicata: essa prevede di determinare anzitutto le inverse $\eta_i$, le loro derivate, e gli intervalli $\varphi([x_{i-1},x_i])$ per cui ciascun addendo $p(\eta_i(y)) \; |\eta_i'(y)|$ contribuisce.

Si osserva che $f(y)$ verifica la condizione di normalizzazione:

$\displaystyle \int_{\varphi([a,b])} f(y)dy = \int_{\varphi([a,b])} \sum_{i=1}^n p(\eta_i(y)) \; |\eta_i'(y)|dy = \sum_{i=1}^n \int_{\varphi([a,b])} p(\eta_i(y)) \; |\eta_i'(y)|dy \;\;$ ma $\eta_i\equiv 0$ al di fuori di $\varphi_i([x_i,x_{i+1}])$!

$\displaystyle \int_{\varphi([a,b])} f(y)dy = \sum_{i=1}^n \int_{\varphi([x_i,x_{i+1}])} p(\eta_i(y)) \; |\eta_i'(y)|dy $

Svolgiamo ora l'integrale per la singola $\varphi_i$, assumendola senza perdere in generalità crescente. Allora $|\eta_i'(y)|=\eta_i'(y); \;\; \varphi_i([x_i,x_{i+1}])=[\varphi_i(x_i),\varphi_i(x_{i+1})]$

$\displaystyle \int_{\varphi_i([x_i,x_{i+1}])} p(\eta_i(y))\; \eta_i'(y)dy = \int_{\varphi_i(x_i)}^{\varphi_i(x_{i+1})} p(\eta_i(y))\; \eta_i'(y)dy = \bigg[ \eta_i(y)\equiv x, dx=\eta_i'(y)dy, y=\varphi_i(x) \bigg] = \int_{x_i}^{x_{i+1}}p(x)dx$

Se $\varphi_i$ fosse stata decrescente il modulo avrebbe cambiato il segno a $\eta_i'$ ma gli estremi di integrazione sarebbero stati invertiti, quindi il risultato sarebbe stato lo stesso. Concludiamo che

$\displaystyle \int_{\varphi([a,b])} = \sum_{i=1}^n \int_{x_i}^{x_{i+1}} p(x)dx = \int_a^b p(x)dx \equiv 1 \;\;\;\;\;\;\;\;\;\;\; \square$

### Un esempio

Supponiamo che $x$ si distribuisca normalmente con valore atteso $\mu$ e varianza $\sigma^2$. Cioè

$\displaystyle p(x)=\frac{1}{\sqrt{2\pi}\sigma} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$

Applichiamo a $x$ la trasformazione

$\displaystyle \varphi(x) = \begin{cases} \dfrac{1}{1+x^2} \;\;\; : \;\; x \leqslant 0 \\ \\ 2e^{-x}-1 \; : \;\; x>0 \end{cases}$

$\varphi$ si può invertire in $(-\infty,0]$ e in $[0,\infty)$:

$\eta_1$ è l'inversa di $\varphi_1 \equiv \frac{1}{1+x^2}$ per $x\leqslant0$:

$\displaystyle y = \frac{1}{1+x^2} \;\; \Rightarrow \;\; y + yx^2 = 1 \;\; \Rightarrow \;\; x^2 = \frac{1-y}{y} \\
\displaystyle \eta_1(y)=-\sqrt{\frac{1-y}{y}} \;\;\; y \in \varphi((-\infty,0])=(0,1] \\
\displaystyle \eta'(y)= \sqrt{\frac{y}{1-y}}\frac{1}{2y^2} > 0$ compatibilmente col fatto che $\varphi_1$ è crescente.

$\eta_2$ è l'inversa di $\varphi_2 \equiv 2e^{-x}-1$:

$\displaystyle y = 2e^{-x}-1 \;\; \Rightarrow \;\; e^{-x}=\frac{y+1}{2} \;\; \Rightarrow \;\; \eta_2(y)=-\log \left (\frac{y+1}{2} \right) \;\;\; y \in \varphi([0,\infty))=(-1,1] \\
\displaystyle \eta_2'(y)=-\frac{1}{y+1} < 0$ compatibilmente col fatto che $\varphi_2$ è decrescente.

Notiamo quindi che, essendo $\varphi((-\infty,0])=[0,1], \; \varphi([0,\infty])=(-1,1]$, per $y \in (-1,0]$ figura solo il secondo contributo alla probabilità; per $y \in [0,1)$ figurano entrambi:

$y \in (-1,0] \;\; \Rightarrow \;\; f(y) = p(\eta_2(y)) \; |\eta_2'(y)| \\
y \in [0,1) \;\;\;\; \Rightarrow \;\; f(y) = p(\eta_1(y)) \; \eta_1'(y) - p(\eta_2(y)) \; \eta_2'(y)$

In definitiva:

$\displaystyle f(y)=\begin{cases} \frac{1}{\sqrt{2\pi}\sigma}\exp \left[-\frac{1}{2\sigma^2} \left( -\log\frac{y+1}{2}-\mu \right)^2 \right] \frac{1}{y+1}  \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; \ : -1<y\leqslant 0 \\ \frac{1}{\sqrt{2\pi}\sigma}\exp \left[-\frac{1}{2\sigma^2} \left( -\log\frac{y+1}{2}-\mu \right)^2\right] \frac{1}{y+1} + \frac{1}{\sqrt{2\pi}\sigma} \exp \left[-\frac{1}{2\sigma^2} \left(-\sqrt{\frac{1-y}{y}}-\mu \right)^2 \right] \sqrt{\frac{y}{1-y}}\frac{1}{2y^2} \;\; : 0\leqslant y < 1 \end{cases}$

.

Osserviamo ora come si distribuisce un campione $\{\varphi(x_i) \}_{i\leqslant N}$ essendo $\{x_i \}_{i\leqslant N}$ un campione gaussiano, e confrontiamo il risultato con la distribuzione teorica $f(y)$.

```python
import numpy as np
import matplotlib.pyplot as plt
import random as ran
e=np.exp
N=10**6
μ,σ = -1,2

def ϕ(x):
    if x<=0: return 1/(1+x**2)
    else:    return 2*e(-x)-1
vec_ϕ=np.vectorize(ϕ)

def f(y):
    if y==-1: return 0.0
    elif -1<y<=0: return 1/(np.sqrt(2*np.pi)*σ) * e( -( np.log((y+1)/2)+μ )**2/(2*σ**2) )/(y+1)
    elif 0<y<1:
        return 1/(np.sqrt(2*np.pi)*σ)* ( e( -( np.log((y+1)/2)+μ )**2/(2*σ**2) )/(y+1) + e(-(np.sqrt((1-y)/y)+μ)**2/(2*σ**2))*np.sqrt(y/(1-y))/(2*y**2))
    elif y==1: return np.inf
vec_f=np.vectorize(f)

X_sample=[]
for i in range(N):
    X_sample.append(np.random.normal(μ,σ))
Y_sample=vec_ϕ(X_sample)
```

```python
X=np.linspace(μ-4*σ,μ+4*σ,2000)
Y=np.linspace(-1,1,800)
u=np.linspace(-10,10,2000)

plt.figure(figsize=(14,8))
plt.title(r'$y=\varphi(x)$',size=20)
plt.plot(u,vec_ϕ(u),'r',lw=3)
plt.grid(c='k',alpha=0.4)
plt.xlabel('$x$',size=18)
plt.ylabel('$y$',size=18)
plt.xlim((-10.1,10.1))
plt.show()

plt.figure(figsize=(14,8))
plt.title('Distribuzione gaussiana: $\mu=$'+str(μ)+', $\sigma=$'+str(σ)+', $N=$'+str(N),size=20)
plt.hist(X_sample,bins=200,density=True,color='b')
plt.plot(X,1/(np.sqrt(2*np.pi)*σ)*e(-(X-μ)**2/(2*σ**2)),c='red',lw=4,alpha=0.6,
         label='$p(x)=\dfrac{1}{\sqrt{2\pi}\sigma}\; e^{-\dfrac{(x-\mu)^2}{2\sigma^2}}$')
plt.xlabel('$x$',size=18)
plt.legend(fontsize=14)
plt.xlim((μ-4.05*σ,μ+4.05*σ))
plt.show()

plt.figure(figsize=(14,8))
plt.title(r'Distribuzione $\varphi(x), \mu=$'+str(μ)+', $\sigma=$'+str(σ)+', $N=$'+str(N),size=20)
plt.hist(Y_sample,bins=200,density=True,color='midnightblue')
plt.plot(Y,vec_f(Y),c='red',lw=4,alpha=0.6,label="$f(y)=p(\eta_1(y))\; |\eta_1'(y)| \;+\; p(\eta_2(y))\; |\eta_2'(y)|$")
plt.xlabel('$x$',size=18)
plt.legend(fontsize=14)
plt.xlim((-1.01,1.01))
plt.show()
```
