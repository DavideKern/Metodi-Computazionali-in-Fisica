### Parte IV: Caduta libera di un grave con attrito

Studiare il moto di un grave con attrito dell'aria descritto dalla forza

$\displaystyle \large |\vec{F}_a| = \frac{1}{2} \, C \, \rho \, A \, v^2$

Graficare l'accelerazione, la velocità e la posizione in funzione del tempo. Considerare come grave un corpo sferico di raggio $R = 8 \, mm$, massa $m = 5 \, g$ e $C = 0.5$.

$ \displaystyle m \vec{a} = m \vec{g} + \vec{F}_a $

$ \displaystyle \hat{j}: m a_y = -mg + \frac{1}{2} \, C \, \rho \, A \, v^2 \, (-sgn \, v) \;\;\;$ ove

$\displaystyle \vec{F}_a \cdot \hat{y} = \frac{1}{2} \, C \, \frac{m}{4/3 \, \pi R^3} \, 4 \pi R^2 \, v^2 \left(-\frac{|v|}{v} \right) = - \frac{1}{2} \, C \, \frac{3m}{R} \, v \, |v| \;\;\;$ da cui

$ \displaystyle \hat{j}: m \ddot{y} = -mg - \frac{3mC}{2R} \, v \, |v|$

$ \displaystyle \ddot{y} = - \frac{3C}{2R} \, \dot{y} \, |\dot{y}| - g$

$ \fbox{$ \begin{cases} \displaystyle \frac{dy}{dt} = v \\ \\ \displaystyle \frac{dv}{dt} = f(t,y,v) = - \frac{3C}{2R} \, v \, |v| - g \\ \\ y(0) = y_0; \; v(0) = v_0 \end{cases} $}$

Notare che a regime ($a_y \equiv 0$) deve risultare

$\displaystyle -\frac{3C}{2R} v_{\infty} \, |v_{\infty}| = g \;\;\; \Rightarrow \;\;\; v_{\infty}^2 = \frac{2Rg}{3C}$

$\displaystyle v_{\infty} = - \sqrt{\frac{2Rg}{3C}} \;\;\;$ - la soluzione positiva per $v_{\infty}$ è chiaramente incompatibile con l'equazione precedente.

### Soluzioni analitiche:

Le soluzioni si trovano dividendo il problema nel caso in cui $|\dot{y}| = \dot{y}$ e quello in cui $|\dot{y}| = -\dot{y}$; le due equazioni che ne conseguono si risolvono come equazioni di Riccati.

Poniamo per comodità $\displaystyle y_0 \equiv y(0), \;\;\; v_0 \equiv v(0), \;\;\; v_{\infty} \equiv - \sqrt{\frac{2Rg}{3C}}, \;\;\; \tau \equiv \sqrt{\frac{2R}{3Cg}}$

 - caso $v_0 \leqslant 0$:
 
     $\displaystyle y(t) = y_0 + v_{\infty}\tau \, \log \bigg( \cosh t/\tau + \frac{v_0}{v_{\infty}} \sinh t/\tau \bigg)$
     
     $\displaystyle v(t) = v_{\infty} \frac{v_0 + v_{\infty} \tanh t/\tau }{v_{\infty} + v_0 \tanh t/\tau} $
     
     $\displaystyle a(t) = -g \; \frac{v_{\infty}^2 - v_0^2}{\Big( v_{\infty} \cosh t/\tau + v_0 \sinh t/\tau \Big)^2} $
     
     In particolare... se $v_0 = v_{\infty}$:
     
     $\displaystyle y(t) = y_0 + v_0 t$
     
     $\displaystyle v(t) = v_0$
     
     $\displaystyle a(t) = \textbf{0}$
     
     
 - caso $v_0 > 0$:
     
     Sia $t_1$ l'istante di inversione del moto ($v(t_1) \equiv 0$); in corrispondenza di esso $\dot{y}$ cambia segno e con esso l'espressione dell'equazione di Newton: le soluzioni andranno riscritte di conseguenza, aggiornando le condizioni iniziali come: $t_0 \equiv t_1, \; y(t_0) \equiv y_1, \; v(t_0) \equiv 0$. In particolare, vale
     
     $\displaystyle t_1 = \tau \tan^{-1} \frac{v_0}{|v_{\infty}|}, \;\;\; y_1 = y_0 - \frac{1}{2} v_{\infty}\tau \, \log \left( 1 + \frac{v_0^2}{v_{\infty}^2} \right), \;\;\; a_1 \equiv a(t_1) = \frac{v_{\infty}}{\tau} = -g$ (comprensibilmente: quando la velocità è nulla, tale è la forza di attrito. Quindi all'accelerazione globale concorre solo la gravità).
     
     $\displaystyle y(t) = \begin{cases} \displaystyle y_0 - v_{\infty}\tau \, \log \bigg( \cos t/\tau - \frac{v_0}{v_{\infty}} \sin t/\tau \bigg) \;\;\; : 0 \leqslant t \leqslant t_1 \\ \\ \displaystyle y_1 + v_{\infty}\tau \, \log \bigg( \cosh \frac{t-t_1}{\tau} \bigg) \;\;\;\;\;\;\;\;\;\;\;\;\;\; : t > t_1 \end{cases}$
     
     $\displaystyle v(t) = \begin{cases} \displaystyle v_{\infty} \frac{v_0 + v_{\infty} \tan t/\tau }{v_{\infty} - v_0 \tan t/\tau} \;\;\; : 0 \leqslant t \leqslant t_1  \\ \\ \displaystyle v _{\infty} \tanh \frac{t-t_1}{\tau} \;\;\;\;\;\;\;\;\; : t > t_1 \end{cases}$
     
     $\displaystyle a(t) = \begin{cases} \displaystyle -g \; \frac{v_{\infty}^2 + v_0^2}{\Big( v_{\infty} \cos t/\tau - v_0 \sin t/\tau \Big)^2} \;\;\; : 0 \leqslant t \leqslant t_1 \\ \\ \displaystyle \DeclareMathOperator{\sech}{sech} -g \; \sech^2 \frac{t-t_1}{\tau} \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\, : t > t_1 \end{cases}$

```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
t=time()

m = .005
R = .008
C = .5
g = 9.81

v_inf = -np.sqrt(2*R*g/(3*C))
τ     =  np.sqrt(2*R/(3*C*g))
α     =  3*C/(2*R)

y0 = 0
v0 = -5*v_inf
t_fin = 5*τ


if v0 <= 0:
    y_an = lambda t: v_inf*τ * np.log(v0/v_inf*np.sinh(t/τ) + np.cosh(t/τ)) + y0
    v_an = lambda t: v_inf   * (v0 + v_inf*np.tanh(t/τ))/(v_inf + v0*np.tanh(t/τ))
    a_an = lambda t: -g * (v_inf**2 - v0**2)/(v_inf*np.cosh(t/τ) + v0*np.sinh(t/τ))**2

else:
    t1 = τ*np.arctan(v0/-v_inf)
    y1 = y0 - v_inf*τ/2 * np.log(1 + v0**2/v_inf**2)

    def y_an(t):
        if t <= t1:
            return   y0 - v_inf*τ * np.log(np.cos(t/τ) - v0/v_inf*np.sin(t/τ))
        else: return y1 + v_inf*τ * np.log(np.cosh((t-t1)/τ))
    def v_an(t):
        if t <= t1:
            return   v_inf * (v0 + v_inf*np.tan(t/τ))/(v_inf - v0*np.tan(t/τ))
        else: return v_inf * np.tanh((t-t1)/τ)
    def a_an(t):
        if t <= t1:
            return   -g * (v_inf**2 + v0**2)/(v_inf*np.cos(t/τ) - v0*np.sin(t/τ))**2
        else: return -g / np.cosh((t-t1)/τ)**2

    y_an = np.vectorize(y_an)
    v_an = np.vectorize(v_an)
    a_an = np.vectorize(a_an)


f  = lambda v: -α*v*abs(v) - g
a0 = f(v0)

def Fehlberg(h = .00001):
    
    global h0, N, t_array, y_array, v_array, a_array
    h0 = h
    N = int(round(t_fin/h, 0))
    
    t_array = np.append([0]    , np.empty(N))
    y_array = np.append([y0]   , np.empty(N))
    v_array = np.append([v0]   , np.empty(N))
    a_array = np.append([f(v0)], np.empty(N))
    
    y_i, v_i = y0, v0
    for i in range(1, N+1):
        t_i = (i-1)*h
        t_array[i] = t_i + h
        
        k1y = h *  v_i
        k2y = h * (v_i + k1y/4)
        k3y = h * (v_i + 3/32*k1y      + 9/32*k2y)
        k4y = h * (v_i + 1932/2197*k1y - 7200/2197*k2y + 7296/2197*k3y)
        k5y = h * (v_i + 439/216*k1y   - 8*k2y         + 3680/513*k3y  - 845/4104*k4y)
        k6y = h * (v_i - 8/27*k1y      + 2*k2y         - 3544/2565*k3y + 1859/4104*k4y - 11/40*k5y)
        
        k1v = h * f(v_i)
        k2v = h * f(v_i + k1v/4)
        k3v = h * f(v_i + 3/32*k1v      + 9/32*k2v)
        k4v = h * f(v_i + 1932/2197*k1v - 7200/2197*k2v + 7296/2197*k3v)
        k5v = h * f(v_i + 439/216*k1v   - 8*k2v         + 3680/513*k3v  - 845/4104*k4v)
        k6v = h * f(v_i - 8/27*k1v      + 2*k2v         - 3544/2565*k3v + 1859/4104*k4v - 11/40*k5v)
        
        y_i += 16/135*k1y + 6656/12825*k3y + 28561/56430*k4y - 9/50*k5y + 2/55*k6y
        v_i += 16/135*k1v + 6656/12825*k3v + 28561/56430*k4v - 9/50*k5v + 2/55*k6v
        
        #y_i += 25/216*k1y + 1408/2565*k3y + 2197/4104*k4y - k5y/5
        #v_i += 25/216*k1v + 1408/2565*k3v + 2197/4104*k4v - k5v/5
        
        y_array[i] = y_i
        v_array[i] = v_i
        a_array[i] = f(v_i)


Fehlberg()

y_an_array = y_an(t_array)
v_an_array = v_an(t_array)
a_an_array = a_an(t_array)

K_array = m*v_array**2/2
U_array = m*g*y_array
E_array = K_array + U_array

K_an_array = m*v_an_array**2/2
U_an_array = m*g*y_an_array
E_an_array = K_an_array + U_an_array


e_y_array = y_array - y_an_array
e_v_array = v_array - v_an_array
e_a_array = a_array - a_an_array
e_K_array = K_array - K_an_array
e_U_array = U_array - U_an_array
e_E_array = E_array - E_an_array

ε_y_array = np.append([0], np.empty(N))
ε_v_array = np.append([0], np.empty(N))
ε_a_array = np.append([0], np.empty(N))
ε_K_array = np.append([0], np.empty(N))
ε_U_array = np.append([0], np.empty(N))
ε_E_array = np.append([0], np.empty(N))

for i in range(1,N+1):
    ε_y_array[i] = e_y_array[i]/y_an_array[i]
    ε_v_array[i] = e_v_array[i]/v_an_array[i]
    ε_a_array[i] = e_a_array[i]/a_an_array[i]
    ε_K_array[i] = e_K_array[i]/K_an_array[i]
    ε_U_array[i] = e_U_array[i]/U_an_array[i]
    ε_E_array[i] = e_E_array[i]/E_an_array[i]


t=time()-t
print('Tempo di esecuzione: t =',round(t,3),'s')
print('h =',h0)
print('N =',N,'\n')

print('τ     =', round(τ, 4), 's')
print('t_fin =', round(t_fin/τ, 4), 'τ =', round(t_fin, 4), 's')
print('v_inf =', round(v_inf, 4), 'm/s')
print('α     =', round(α, 4), 'm⁻¹\n')

if v0 > 0:
    print('t1 =', round(t1/τ,4), 'τ =', round(t1,4), 's')
    print('y1 = y0 +', round(y1-y0,4), 'm =', round(y1,4), 'm')
```

```python
plt.figure(figsize=(16,9))
ax = plt.axes()
ax.set_facecolor('gainsboro')
plt.title(r'$y(t); h = '+str(round(h0,6))+', \; N = '+str(N)+'$ [Fehlberg]', size=22)
plt.text(s=r'$y(0) =' + str(round(y0,3)) + '\; m$', x=t_fin/70, y=y0, color='g', size=14)

plt.plot(t_array, y_array, 'r', lw=3.5, label=r'$y(t)$ - sol. numerica')
plt.plot(t_array, y_an_array, '--',c='yellow', lw=1.5, label=r'$y(t)$ - sol. analitica')
plt.plot([0],[y0],'o g', ms=9, mec='k')
if v0 > 0:
    plt.axvline(t1, c='k', lw=1, label=r'$t_1 = ' + str(round(t1,3)) + '\; s$')
    plt.plot([t1],[y1],'o r', ms=9, mec='k')
    plt.text(s=r'$y_1 =' + str(round(y1,3)) + '\; m$', x = t1 + t_fin/20, y = y1, color='r', size=14)

plt.xlabel('$t \; [s]$', size=20)
plt.ylabel('$y \; [m]$', size=20)
plt.legend(fontsize=16)
plt.show()



plt.figure(figsize=(16,9))
ax = plt.axes()
ax.set_facecolor('gainsboro')
plt.title(r'$v(t); h = '+str(round(h0,6))+', \; N = '+str(N)+'$ [Fehlberg]', size=22)
plt.text(s=r'$v(0) =' + str(round(v0,3)) + '\; m/s$', x=t_fin/70, y=v0, color='g', size=14)
plt.text(s=r'$v_{\infty} = ' + str(round(v_inf,3)) + '\; m/s$', x=0, y=.95*v_inf, color='k', size=14)

plt.axhline(v_inf, c='k', ls='--', lw=1)
plt.plot(t_array, v_array, 'b', lw=3.5, label=r'$v(t)$ - sol. numerica')
plt.plot(t_array, v_an_array, '--',c='yellow', lw=1.5, label=r'$v(t)$ - sol. analitica')
plt.plot([0],[v0],'o g', ms=9, mec='k')
if v0 > 0:
    plt.axvline(t1, c='k', lw=1, label=r'$t_1 = ' + str(round(t1,3)) + '\; s$')
    plt.plot([t1],[0],'o b', ms=9, mec='k')
    plt.text(s=r'$v_1 = 0 \; m/s$', x = t1 + t_fin/70, y = v0/30, color='b', size=14)
    
plt.xlabel('$t \; [s]$', size=20)
plt.ylabel('$v_y \; [m/s]$', size=20)
plt.legend(fontsize=16)
plt.show()



plt.figure(figsize=(16,9))
ax = plt.axes()
ax.set_facecolor('gainsboro')
plt.title(r'$a(t); h = '+str(round(h0,6))+', \; N = '+str(N)+'$ [Fehlberg]', size=22)
plt.text(s=r'$a(0) =' + str(round(a0,3)) + '\; m/s^2$', x=t_fin/70, y=a0, color='g', size=14)

plt.axhline(0, c='k', ls='--', lw=1)
plt.plot(t_array, a_array, 'indigo', lw=3.5, label=r'$a(t)$ - sol. numerica')
plt.plot(t_array, a_an_array, '--',c='yellow', lw=1.5, label=r'$a(t)$ - sol. analitica')
plt.plot([0],[a0],'o g', ms=9, mec='k')
if v0 > 0:
    plt.axvline(t1, c='k', lw=1, label=r'$t_1 = ' + str(round(t1,3)) + '\; s$')
    plt.plot([t1],[-g], 'indigo', marker='o', ms=9, mec='k')
    plt.text(s=r'$a_1 = -9.81 \; m/s^2$', x = t1 + t_fin/50, y = -g - abs(a0+g)/20, color='indigo', size=14)

plt.xlabel('$t \; [s]$', size=20)
plt.ylabel('$a_y \; [m/s^2]$', size=20)
plt.legend(fontsize=16)
plt.show()
```

### Verifica del teorema dell'energia meccanica

Nel moto di caduta con attrito è chiaro che l'energia meccanica $E_M$ non si conservi: $\vec{F}_a$ è una forza non conservativa che compie lavoro negativo, quindi contribuisce a far diminuire l'energia meccanica del sistema:

$\displaystyle E_M (t) \equiv K(t) + U(t) = m \, \left( \frac{v^2(t)}{2} + g y(t) \right) \\ \displaystyle \vec{F}_a \cdot \hat{j} = -\frac{3mC}{2R} v \, |v| \equiv -\alpha m \, v \, |v|$
 
Essendo nota l'espressione della forza d'attrito, è anche possibile trovarne una per il suo lavoro. In definitiva, si può verificare se vale il ***teorema dell'energia meccanica***:

$\displaystyle \Delta E_M = \mathcal{L}_{nc} \;\;\; \Leftrightarrow \;\;\; E_M(t) - E_M(0) = \mathcal{L}_{nc}(t) \;\;\; \Leftrightarrow \;\;\; E_M(0) = E_M(t) - \mathcal{L}_{nc}(t) = cost. \equiv m \, \left( \frac{v_0^2}{2} + g y_0 \right)$

essendo $\displaystyle \mathcal{L}_{nc}(t) = \int_0^t \vec{F}_{a}(t') \cdot \vec{v}(t') \; dt' = -\alpha m \int_0^t |v^3(t')| \; dt'$ il lavoro compiuto dall'attrito fino all'istante $t$. Per testare la validità del teorema si possono rappresentare $E_M(t), -\mathcal{L}_{nc}(t)$ (quantità positiva) e la differenza $E_M(t) - \mathcal{L}_{nc}(t)$, per verificare se questa si mantiene uguale a $E_M(0)$.

Anche se $|v^3(t)|$ non è analiticamente integrabile, è comunque nota per punti; pertanto se ne può stimare la primitiva applicando iterativamente il metodo dei trapezi:

$\mathcal{L}_{nc}(0) \equiv 0 \\ \displaystyle \mathcal{L}_{nc}(t_k) = \left( \int_0^{t_{k-1}} + \int_{t_{k-1}}^{t_k} \right) \left[ -\alpha m \; |v^3|(t') \right] \; dt' = \mathcal{L}_{nc}(t_{k-1}) + \int_{t_{k-1}}^{t_k} - \alpha m \; |v^3|(t') \; dt' \sim \mathcal{L}_{nc}(t_{k-1}) - \alpha m \; \frac{t_{k} - t_{k-1}}{2} \; \left( |v^3(t_k)| + |v^3(t_{k-1})| \right) $

ossia:

$\mathcal{L}_{nc}(0) \equiv 0 \\ \displaystyle \mathcal{L}_{nc}(t_k) \sim \mathcal{L}_{nc}(t_{k-1}) - \alpha m \; \frac{h}{2} \; \left( |v^3(t_k)| + |v^3(t_{k-1})| \right) \;\;\;\; t_k = \frac{k}{N}T, \;\; k \in \{1, \cdots, N \}$

Nel programma, per comodità, è stata prima stimata la primitiva di $|v^3(t)|$ e dopo, moltiplicando per $-\alpha m$, l'array di $\mathcal{L}_{nc}(t)$.

```python
E0 = m*(v0**2/2 + g*y0)

Lnc_array    = np.append([0], np.empty(N))
Lnc_an_array = np.append([0], np.empty(N))

for j in range(1,N+1):
    Lnc_array[j]    = Lnc_array[j-1]    + h0/2 * (abs(v_array[j-1])**3    + abs(v_array[j])**3)
    Lnc_an_array[j] = Lnc_an_array[j-1] + h0/2 * (abs(v_an_array[j-1])**3 + abs(v_an_array[j])**3)

Lnc_array    *= -α*m
Lnc_an_array *= -α*m



plt.figure(figsize=(16,9))
ax = plt.axes()
ax.set_facecolor('gainsboro')
plt.title(r'$K(t), U(t), E_M(t)$ [Fehlberg]', size=22)

plt.axhline(0, c='w', lw=1)
plt.plot(t_array, K_array, 'green',     lw=3.5, label=r'$K(t)$   -  sol. numerica')
plt.plot(t_array, U_array, 'blue',      lw=3.5, label=r'$U(t)$   -  sol. numerica')
plt.plot(t_array, E_array, 'goldenrod', lw=3.5, label=r'$E_M(t)$ -  sol. numerica')
plt.plot(t_array, U_an_array, '--k',  lw=1,  label=r'$K, U, E_M$ -  sol. analitiche')
plt.plot(t_array, K_an_array, '--k',  lw=1)
plt.plot(t_array, E_an_array, '--k',  lw=1)
if v0 > 0:
    plt.axvline(t1, c='k', lw=1, label=r'$t_1 = ' + str(round(t1,3)) + '\; s$')

plt.xlabel('$t  \; [s]$', size=20)
plt.ylabel('$K,U,E_M \;\; [\,J\,]$', size=20)
plt.legend(fontsize=16, loc='upper right')
plt.xlim(-.005*t_fin,t_fin)
plt.show()


plt.figure(figsize=(16,9))
ax = plt.axes()
ax.set_facecolor('gainsboro')
plt.title(r'$E_M(t), -\mathcal{L}_{nc}(t), E_M(0)='+str(round(E0,3))+'\; J$ [Fehlberg]', size=22)

plt.axhline(0, c='w', lw=1)
plt.plot(t_array, E_array,   'goldenrod',  lw=3.5, label=r'$E_M(t)$ - sol. numerica')
plt.plot(t_array, -Lnc_array,'darkorange', lw=3.5, label=r'$-\mathcal{L}_{nc}(t)$ - sol. numerica')
plt.plot(t_array, E_array-Lnc_array, 'y',  lw=3.5, label=r'$E_M(t) - \mathcal{L}_{nc}(t)$ - sol. numerica')
plt.plot(t_array, E_an_array,  '--k',      lw=1,   label=r'$E_M, -\mathcal{L}_{nc}, E_M(0)$ - sol. analitiche')
plt.plot(t_array, -Lnc_an_array, '--k',    lw=1)
plt.axhline(E0, c='k', ls='--',lw=1)
if v0 > 0:
    plt.axvline(t1, c='k', lw=1, label=r'$t_1 = ' + str(round(t1,3)) + '\; s$')

plt.xlabel('$t  \; [s]$', size=20)
plt.ylabel('$E_M, \; -\mathcal{L}_{nc}, \; E_M(0) \; [\,J\,]$', size=20)
plt.legend(fontsize=16, loc='center right')
plt.xlim(-.005*t_fin,t_fin)
plt.show()


plt.figure(figsize=(16,9))
ax = plt.axes()
ax.set_facecolor('gainsboro')
plt.title(r'$E_M(t) - \mathcal{L}_{nc}(t)$ [Fehlberg]', size=22)

plt.plot(t_array, E_array-Lnc_array, 'y', lw=3.5, label=r'$E_M(t) - \mathcal{L}_{nc}(t)$ - sol. numerica')
plt.axhline(E0, c='k', ls='--',lw=1)

plt.xlabel('$t  \; [s]$', size=20)
plt.ylabel('$E_ - \mathcal{L}_{nc} \; [\,J\,]$', size=20)
plt.legend(fontsize=16)
plt.xlim(-.005*t_fin,t_fin)
plt.show()
```
