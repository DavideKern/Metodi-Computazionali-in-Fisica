### Parte IV: Moto nel Campo Magnetico e Elettrico (con attrito)

Studiare il moto di un elettrone in un campo magnetico ortogonale al vettore velocità iniziale. L'equazione del moto è

$\Large \displaystyle m \ \frac{d \vec{v}}{dt} = q \; \left(\vec{v} \times \vec{B} + \vec{E} \right) - \beta \; \vec{v}$

Calcolare numericamente e rappresentare la traiettoria dell'elettrone con un campo magnetico omogeneo, costante e ortogonale alla velocità iniziale. Partire dal caso $\vec{E} = \textbf{0}$; successivamente variare $\vec{B}$. Studiare poi gli stessi casi aggiungendo un campo elettrico uniforme e costante, anch'esso ortogonale al campo magnetico.

$ \displaystyle \vec{B} \equiv (0, \; 0, \; B), \;\;\;\; \vec{E} \equiv (E_x, \; E_y, \; 0)$

$\displaystyle \vec{r}(0) \equiv (x_0, \; y_0, \; 0), \;\;\;\; \vec{v}(0) \equiv \left( v_x^0, \; v_y^0, \; 0 \right)$

Svolgendo i calcoli, si trova che le equazioni del moto in forma normale sono

$\begin{cases} \dot{x} = v_x \\ \\ \dot{y} = v_y \\ \\ \dot{v}_x = -\gamma v_x + \omega v_y - \omega v_y^c \\ \\ \dot{v}_y = -\omega v_x - \gamma v_y + \omega v_x^c \end{cases}$

avendo posto

$\displaystyle \omega \equiv \frac{qB}{m}$ ($\omega$ ha un segno che dipende dalla carica $q$ e dalla proiezione lungo l'asse $z$ di $\vec{B}$)

$\displaystyle \gamma \equiv \frac{\beta}{m}$

$\displaystyle v_x^c \equiv \frac{E_y}{B}; \;\;\; v_y^c \equiv -\frac{E_x}{B}$

Il sistema, ristretto all'incognita $\underline{v}(t) \equiv (v_x(t), v_y(t))^t$, può essere riespresso mediante il formalismo matriciale come

$\displaystyle \underline{\dot{v}}(t) = A \cdot \underline{v}(t) + \omega \; \begin{pmatrix} -v_y^c \\ v_x^c \end{pmatrix} \;\;\;\;$ ove $\displaystyle A \equiv \begin{pmatrix} -\gamma & \omega \\ -\omega & -\gamma \end{pmatrix}$

Si tratta di un ***sistema differenziale lineare del prim'ordine in due incognite*** e con termine noto. L'integrale generale è la somma dell'integrale generale dell'equazione omogenea associata ($\displaystyle \underline{\dot{v}} = A \cdot \underline{v}$) con un integrale particolare dell'equazione completa.

Come integrale particolare della completa possiamo scegliere la soluzione costante: $\displaystyle \underline{v}^{\star} \equiv (v_x^{\star}, v_y^{\star})$. Sostituendo nel sistema iniziale, ricaviamo un sistema lineare nelle due incognite $v_x^{\star}$ e $v_y^{\star}$. È facile mostrare che risulta

$\begin{cases} \displaystyle v_x^{\star} = \frac{v_x^c - \gamma/\omega \; v_y^c}{1 + (\gamma/\omega)^2} = \frac{1}{B} \frac{\frac{\beta}{qB} E_x + E_y}{1 + \left( \frac{\beta}{qB} \right)^2} \\ \\
\displaystyle v_y^{\star} = \frac{\gamma/\omega \; v_x^c + v_y^c}{1 + (\gamma/\omega)^2} = \frac{1}{B} \frac{-E_x + \frac{\beta}{qB}E_y}{1 + \left( \frac{\beta}{qB} \right)^2} \end{cases}$

Notiamo che la soluzione, nel caso $\beta = 0$, si riduce a $\underline{v}^{\star} = \underline{v}_c = (v_x^c, \; v_y^c)^t$.

La soluzione generale dell'omogenea associata è $\displaystyle e^{A t} \cdot \underline{c}$, ove $\displaystyle e^{A t} \equiv \sum_{k=0}^{\infty} \frac{A^k t^k}{k!}$ e $\displaystyle \underline{c} \equiv (c_x, c_y)^t$ è un vettore di costanti da fissare con il dato iniziale.

Nel nostro caso, [si potrebbe dimostrare](http://www.mat.uniroma2.it/~digennar/EqDiff.pdf) che vale

$\displaystyle e^{At} = e^{-\gamma t} \begin{pmatrix} \cos \omega t & \sin \omega t \\ -\sin \omega t & \cos \omega t \end{pmatrix} = e^{-\gamma t} \, \mathcal{R}_{\omega}(t)$

In definitiva

$\displaystyle v_x(t) = e^{-\gamma t} \left[ c_x \cos \omega t + c_y \sin \omega t \right] + v_x^{\star} \\
 \displaystyle v_y(t) = e^{-\gamma t} \left[ -c_x \sin \omega t + c_y \cos \omega t \right] + v_y^{\star}$

E imponendo $\underline{v}(0) \equiv \underline{v}_0 = (v_x^0, \; v_y^0)^t$, si ricava $c_x = v_x^0 - v_x^{\star}; \;\; c_y = v_y^0 - v_y^{\star}$. Da cui

$\displaystyle v_x(t) = e^{-\gamma t} \left[ (v_x^0 - v_x^{\star}) \cos \omega t + (v_y^0 - v_y^{\star}) \sin \omega t \right] + v_x^{\star} \\
 \displaystyle v_y(t) = e^{-\gamma t} \left[ (v_y^0 - v_y^{\star}) \cos \omega t - (v_x^0 - v_x^{\star}) \sin \omega t \right] + v_y^{\star}$

Nel formalismo matriciale, la soluzione si può esprimere come

$\displaystyle \underline{v}(t) = \underline{v}^{\star} + e^{-\gamma t} \mathcal{R}_{\omega}(t) \cdot [\underline{v}_0 - \underline{v}^{\star}]$

Il vettore velocità, partendo dal valore $\vec{v}_0$, ruota attorno l'offset $\vec{v}^{\star}$ con un'ampiezza che si attenua esponenzialmente: a regime risulterà quindi $\vec{v}_{\infty} = \vec{v}^{\star}$

A campo elettrico spento, risulta $\vec{v}^{\star} = \textbf{0}$ e la particella, a regime, si ferma (comprensibilmente, venendo a mancare la forzante).

In generale, ci aspettiamo che la particella si sposti, spiraleggiando, nella direzione di $\vec{v}^{\star}$; il raggio delle oscillazioni si smorzerà esponenzialmente secondo la costante di tempo $\tau \equiv 1/\gamma = m/\beta$: maggiore sarà la costante di attrito $\beta$, minore sarà $\tau$ e più rapido sarà il raggiungimento del regime: la particella raggiungerà, nel limite, lo stato di moto rettilineo uniforme.

È interessante notare che, a differenza del caso privo di attrito, **$\vec{v}^{\star}$ non è ortogonale al campo elettrico**. Con i calcoli non è difficile dimostrare che

- $\displaystyle \vec{v}^{\star} \cdot \vec{E} = \frac{E^2}{B} \; \frac{(\beta \ / \ qB)}{1 + (\beta \ / \ qB)^2}$



- $\displaystyle v^{\star} \equiv |\vec{v}^{\star}| = \frac{|E|}{|B|} \; \frac{1}{\sqrt{1 + (\beta \ / \ qB)^2}}$

Sapendo che vale $\displaystyle \cos \theta = \frac{\vec{v}^{\star} \cdot \vec{E}}{v^{\star}|E|}$ è possibile ricavare l'angolo tra la velocità di regime e il campo:

$\displaystyle \theta = \arccos \, \left( \frac{ \frac{\beta}{q|B|} }{\sqrt{1 + \left( \frac{\beta}{q|B|} \right)^2}} \right)$

$\theta$ *non dipende* dall'intensità di $\vec{E}$, né dalla sua direzione.

 - per $\beta=0$ ritroviamo lo scenario in assenza di attrito: $\cos \theta = 0 \;\; \Rightarrow \;\; \theta = \pi/2 \;\; \Rightarrow \;\; \vec{v}^{\star} \perp \vec{E}$;
 - per $\beta \gg |qB|$ si ha $\cos \theta \simeq 1 \;\; \Rightarrow \;\; \theta \simeq 0 \;\; \Rightarrow \;\; \vec{v}^{\star} \parallel \vec{E}$. È il caso in cui l'attrito è molto elevato rispetto l'interazione fra la particella e il campo magnetico;
 - per $\beta = q|B|$ si ha una situazione intermedia: infatti $\cos \theta = 1/\sqrt{2}$, e $\vec{v}^{\star}$ forma un angolo di 45° gradi con $\vec{E}$. Nel caso dell'elettrone, essendo $\beta$ definito positivo, risulterà $\cos \theta = -1/\sqrt{2}$ e $\theta = \pi - \pi/4 = 3\pi/4 = 135°$.

Questa trattazione fa capire che per analizzare il comportamento del sistema, conviene variare $\beta$ *non* in assoluto, ma in termini di multipli di $|qB|$. Per questo motivo, nel programma essa è inizializzata come ```β = α * abs(q*B)```.

Sapendo che

$\displaystyle \int e^{-\gamma t} \cos \omega t \; dt = e^{-\gamma t} \; \frac{\omega \sin \omega t - \gamma \cos \omega t}{\omega^2 + \gamma^2}; \;\;\;\; \int e^{-\gamma t} \sin \omega t \; dt = -e^{-\gamma t} \; \frac{\gamma \sin \omega t + \omega \cos \omega t}{\omega^2 + \gamma^2}$

è possibile ricavare le leggi orarie come

$\displaystyle x(t) = x_0 + \int_0^t v_x(t') \; dt' \;\;\;\; y(t) = y_0 + \int_0^t v_y(t') \; dt'$

In particolare, risulta

$\begin{cases} \displaystyle x(t) = x_0 + \frac{(v_x^0 - v_x^{\star}) \gamma + (v_y^0 - v_y^{\star}) \omega}{\omega^2 + \gamma^2} + v_x^{\star}t - \frac{e^{-\gamma t}}{\omega^2 + \gamma^2} \bigg\{ \Big[(v_x^0 - v_x^{\star})\gamma + (v_y^0 - v_y^{\star})\omega \Big]\cos \omega t + \Big[(v_y^0 - v_y^{\star})\gamma - (v_x^0 - v_x^{\star})\omega \Big] \sin \omega t \bigg\} \\ \\
\displaystyle y(t) = y_0 + \frac{(v_y^0 - v_y^{\star}) \gamma - (v_x^0 - v_x^{\star}) \omega}{\omega^2 + \gamma^2} + v_y^{\star}t - \frac{e^{-\gamma t}}{\omega^2 + \gamma^2} \bigg\{ \Big[(v_y^0 - v_y^{\star})\gamma - (v_x^0 - v_x^{\star})\omega \Big]\cos \omega t - \Big[(v_x^0 - v_x^{\star})\gamma + (v_y^0 - v_y^{\star})\omega \Big] \sin \omega t \bigg\} \end{cases}$

Per contrarre la notazione possiamo porre

$\begin{cases} \displaystyle x_c \equiv x_0 + \frac{(v_x^0 - v_x^{\star})\gamma + (v_y^0 - v_y^{\star})\omega }{\omega^2 + \gamma^2} \\ \\
\displaystyle y_c \equiv y_0 + \frac{(v_y^0 - v_y^{\star})\gamma - (v_x^0 - v_x^{\star})\omega }{\omega^2 + \gamma^2} \end{cases}$

ottenendo

$\displaystyle x(t) = x_c + v_x^{\star}t + e^{-\gamma t} \Big[(x_0 - x_c) \cos \omega t + (y_0 - y_c) \sin \omega t \Big] \\ \\
\displaystyle y(t) = y_c + v_y^{\star}t + e^{-\gamma t} \Big[(y_0 - y_c) \cos \omega t - (x_0 - x_c) \sin \omega t \Big]$

Nel formalismo matriciale risulta

$\displaystyle \underline{r}(t) = \underline{r}_c + \underline{v}^{\star} t + e^{-\gamma t} \; \mathcal{R}_{\omega}(t) \cdot \left( \underline{r}_0 - \underline{r}_c \right) \;\;\;\;\;\; \mathcal{R}_{\omega}(t) = \begin{pmatrix} \cos \omega t & \sin \omega t \\ -\sin \omega t & \cos \omega t \end{pmatrix}$

Geometricamente vuol dire che la particella, partendo da $\vec{r}_0$, inizia a ruotare attorno la posizione $\vec{r}_c$; contemporaneamente $\vec{v}^{\star}$ agisce da velocità di trascinamento, combinando un moto circolare smorzato con un moto rettilineo uniforme. Ci aspettiamo che la traiettoria sia una spirale piana con raggio via via decrescente (per via degli attriti). Nel caso semplice $\vec{E} = \textbf{0}$ abbiamo già visto che $\vec{v}^{\star} = \textbf{0}$, per cui il moto si riduce a circolare smorzato attorno $\vec{r}_c$. Ossia

$\displaystyle \lim_{t \rightarrow \infty} \vec{r}(t) = \vec{r}_c$

In generale, il raggio iniziale $R_0$ della traiettoria sarà

$\displaystyle R_0^2 = \left( x_c - x_0 \right)^2 + \left( y_c - y_0 \right)^2 = \frac{(v_x^0 - v_x^{\star})^2 + (v_y^0 - v_y^{\star})^2}{\omega^2 + \gamma^2}$

Nel caso privo di attrito ($\gamma = 0$) e a campo elettrico spento ($\vec{v}^{\star} = \textbf{0}$) risulterà

$\displaystyle R_0^2 = \frac{\left( v_x^0 \right)^2 + \left( v_y^0 \right)^2}{\omega^2} = \frac{|v_0|^2}{(qB)^2 \,/ \, m^2} \;\;\; \Rightarrow \;\;\; R_0 = \frac{m v_0}{|qB|}$: abbiamo ritrovato il raggio di Larmor.

Le accelerazioni si ottengono facilmente derivando $v_x(t)$ e $v_y(t)$. Mettendo in evidenza l'esponenziale e i due termini $\cos \omega t, \sin \omega t$, si ottiene ancora una volta

$\begin{cases} \displaystyle a_x(t) = e^{-\gamma t} \left[ a_x^0 \cos \omega t + a_y^0 \sin \omega t \right] \\ \\
\displaystyle a_y(t) = e^{-\gamma t} \left[ a_y^0 \cos \omega t - a_x^0 \sin \omega t \right] \end{cases}$

avendo posto

$\displaystyle a_x^0 \equiv -\gamma(v_x^0 - v_x^{\star}) + \omega (v_y^0 - v_y^{\star}) \\ \\
\displaystyle a_y^0 \equiv -\gamma(v_y^0 - v_y^{\star}) - \omega (v_x^0 - v_x^{\star})$

In realtà, ricordando le equazioni del moto e osservando che $a_x^0, a_y^0$ coincidono con $\dot{v}_x(0), \dot{v}_y(0)$, si possono usare anche le espressioni più semplici

$\begin{cases} \displaystyle a_x^0 = -\gamma v_x^0 + \omega v_y^0 + q \frac{E_x}{m} \\ \\ \displaystyle a_y^0 = -\omega v_x^0 - \gamma v_y^0 + q \frac{E_y}{m} \end{cases}$

Nel formalismo matriciale risulta

$\displaystyle \underline{a}(t) = e^{-\gamma t} \; \mathcal{R}_{\omega}(t) \cdot \underline{a}_0$

L'accelerazione è un vettore ruotante attorno l'origine, e si smorza esponenzialmente proprio come la componente ruotante di $\vec{v}(t)$ e $\vec{r}(t)$. A differenza di questi ultimi, però, l'accelerazione non ha alcun offset: a regime essa tende sempre ad annullarsi. Questo aspetto è compatibile col fatto che il moto della particella, per $t \gg \tau$, è praticamente rettilineo uniforme.

### Considerazioni energetiche

L'energia meccanica non si conserva per via dell'attrito. Possiamo tuttavia verificare la validità del ***teorema dell'energia meccanica***:

$\displaystyle \Delta \mathcal{E}_M = \mathcal{L}_{nc} \;\;\; \Leftrightarrow \;\;\; \mathcal{E}_M(t) - \mathcal{E}_M(0) = \mathcal{L}_{nc}(t) \;\;\; \Leftrightarrow \;\;\; \mathcal{E}_M(0) = \mathcal{E}_M(t) - \mathcal{L}_{nc}(t) = cost. \equiv \frac{1}{2}mv_0^2 \;\;\;$ (facendo la scelta $U(0) = U(\vec{r}_0) \equiv 0$).

Siano $\displaystyle K(t) \equiv \frac{1}{2}mv^2(t)$ e $\displaystyle U(t) \equiv -q \; \big[ E_x(x(t) - x_0) + E_y(y(t) - y_0) \big]$ le energie cinetica e potenziale del sistema. Notiamo che essendo

$\displaystyle \lim_{t\rightarrow \infty} \vec{v}(t) = \vec{v}^{\star}$

risulterà anche

$\displaystyle \lim_{t \rightarrow \infty} K(t) \equiv K_{\infty} = \frac{1}{2}mv^{\star \, 2}$

Invece, l'andamento asintotico di $\vec{r}(t)$ - un moto rettilineo uniforme - si ottiene eliminando il transiente ($ e^{-\gamma t} \, \mathcal{R}_{\omega}(t) \cdot \left[ \vec{r}_0 - \vec{r}_c \right]$):

$\displaystyle \vec{r}_{\infty}(t) \equiv \vec{r}_c + \vec{v}^{\star}t$

Sostituendo in $U(t)$ e ricordando l'espressione di $x_c, y_c$, ricaviamo l'andamento asintotico del potenziale:

$U_{\infty}(t) = U_1 + P_{\infty}t \;\;$ avendo posto

$\displaystyle U_1 \equiv -q \; \left[ E_x \; \frac{(v_x^0 - v_x^{\star})\gamma + (v_y^0 - v_y^{\star})\omega}{\omega^2+\gamma^2} + E_y \; \frac{(v_y^0 - v_y^{\star})\gamma - (v_x^0 - v_x^{\star})\omega}{\omega^2+\gamma^2} \right]$

$\displaystyle P_{\infty} = -q \; \left[ E_x \ v_x^{\star} + E_y \ v_y^{\star} \right] = -q \; \vec{E} \cdot \vec{v}^{\star}$

Cioè, a regime $K(t)$ si stabilizza a un valore limite, mentre $U(t)$ continua a variare con un andamento lineare (comprensibilmente: in presenza di attrito la velocità di trascinamento $\vec{v}^{\star}$ non è più ortogonale al campo, quindi la particella non avanza lungo una superficie equipotenziale).

Ma è vero che l'energia meccanica diminuisce nel tempo? Sostituendo i valori asintotici in $\mathcal{E}(t)$, abbiamo che

$\displaystyle \mathcal{E}_{\infty}(t) = K_{\infty} + U_{\infty}(t) = K_{\infty} + U_1 + P_{\infty}t$ da cui

$\displaystyle \dot{\mathcal{E}}_{\infty} = P_{\infty}$

Sostituendo in $P_{\infty}$ i valori di $v_x^{\star}$ e $v_y^{\star}$, otteniamo

$\displaystyle P_{\infty} = -\frac{E^2}{B^2} \; \frac{\beta}{(\beta/qB)^2 + 1} < 0$

Quindi, compatibilmente con la fisica, $\exists \, t'>0 \; t.c.: \forall t > t' \;\; \displaystyle \frac{d\mathcal{E}(t)}{dt} < 0$

$\displaystyle P_{\infty} = -q \; \vec{v}_c \cdot \vec{E} = 0$ essendo $\vec{v}_c \perp \vec{E}$.

$\displaystyle \mathcal{L}_{nc}(t) = \int_0^t \vec{f}_{av}(t') \cdot \vec{v}(t') \; dt' = \int_0^t -\beta \; v^2(t') \; dt'$

$\displaystyle \mathcal{L}_{\infty}(t) = \mathcal{L}_1 + P_{\infty} t$ ove

````python
t=time()

m        = me                  #massa della particella
q        = -e                  #carica della particella
B        = +2.5e-6             #componente lungo z del campo magnetico in T
β        = 0.05*abs(q*B)       #costante di attrito viscoso in kg/s
Ex, Ey   = +1.5e-2, -2.0e-2    #componenti lungo x,y del campo elettrico in V/m
x0,  y0  =  4, -3              #posizione iniziale in m
vx0, vy0 = +3.0e+5, +4.0e+5    #velocità iniziale in m/s

γ  = β/m
τ  = m/β
T  = 2*np.pi*m/abs(q*B)        #periodo di Larmor in s
ω  = q*B/m                     #pulsazione in rad/s (il segno di ω indica
E  = np.sqrt(Ex**2 + Ey**2)    #se la rotazione avviene in senso orario o antiorario)

v0      = np.sqrt(vx0**2 + vy0**2)               #modulo iniziale della velocità
vx_star = (β/(q*B)*Ex + Ey)/((β/(q*B))**2 + 1)/B #valore a regime della velocità
vy_star = (β/(q*B)*Ey - Ex)/((β/(q*B))**2 + 1)/B
v_star  = E/abs(B)/np.sqrt(1 + (β/(q*B))**2)
V       = np.sqrt((vx0 - vx_star)**2 + (vy0 - vy_star)**2)                 #ampiezza di oscillazione iniziale della velocità
x_c     = ((vx0 - vx_star)*γ + (vy0 - vy_star)*ω)/(ω**2 + γ**2) + x0       #centro iniziale iniziale di oscillazione
y_c     = ((vy0 - vy_star)*γ - (vx0 - vx_star)*ω)/(ω**2 + γ**2) + y0
R0      = np.sqrt(((vx0 - vx_star)**2 + (vy0 - vy_star)**2)/(ω**2 + γ**2)) #raggio iniziale di oscillazione
θ       = np.arccos(β/(q*abs(B))/np.sqrt(1 + (β/(q*abs(B)))**2))           #angolo fra velocità a regime e campo elettrico

u1 = vx_star*vx0 + vy_star*vy0 - v_star**2
u2 = vx_star*vy0 - vy_star*vx0

ax0 = -γ*vx0 + ω*vy0 + q*Ex/m  #accelerazione iniziale in m/s^2
ay0 = -γ*vy0 - ω*vx0 + q*Ey/m
a0  = np.sqrt(ax0**2 + ay0**2) #modulo dell'accelerazione iniziale

t_fin = 10*τ                    #duranta temporale da indagare (la si può porre ad esempio a τ/2, τ, 2τ, 3τ, ...)
Fehlberg(t_fin/100000)


EM_0  = m/2 * v0**2                           #energia meccanica/cinetica iniziale
K_inf = m/2 * v_star**2                       #energia cinetica a regime
P_inf = -E**2/B**2 * β/( (β/(q*B))**2 + 1 )   #pendenza a regime dell'energia potenziale
U_1   = -q * (Ex*(x_c - x0) + Ey*(y_c - y0))  #offset a regime dell'energia potenziale
L_1   = -m/2 * ((vx0 - vx_star)**2 + (vy0 - vy_star)**2) - 2*β/(ω**2 + γ**2)*(γ*u1 + ω*u2) #offset a regime del lavoro non conservativo

odg_t = int(np.floor(np.log10(t_fin)))
odg_v = 0
odg_a = 0
if V>0:
    odg_v = int(np.log10(V))
    odg_a = int(np.log10(V*abs(ω)))


x_an_array  = x_an(t_array)
y_an_array  = y_an(t_array)
vx_an_array = vx_an(t_array)
vy_an_array = vy_an(t_array)
ax_an_array = ax_an(t_array)
ay_an_array = ay_an(t_array)

K_array    = m/2 * (vx_array**2 + vy_array**2)
U_array    = -q * (Ex*(x_array-x0) + Ey*(y_array-y0))
EM_array   = K_array + U_array

K_an_array   = m/2 * (vx_an_array**2 + vy_an_array**2)
U_an_array   = -q * (Ex*(x_an_array-x0) + Ey*(y_an_array-y0))
EM_an_array  = K_an_array + U_an_array
Lnc_an_array = Lnc(t_array)


t=time()-t
print('Tempo di esecuzione: t =',"{0:.4g}".format(t),'s\n')
print('h     =',"{0:.4g}".format(h0),'s')
print('N     =',N)
print('t_fin =',"{0:.4g}".format(t_fin),'s =',"{0:.4g}".format(t_fin/T),'T =',"{0:.4g}".format(t_fin/τ),'τ\n')

print('m   =', "{0:.4g}".format(m),'kg')
print('q   =', "{0:.4g}".format(q),'C')
print('β   =', "{0:.4g}".format(β),'kg/s =',"{0:.4g}".format(β/abs(q*B)),'|qB|')
print('B_z =', "{0:.4g}".format(B),'T')
print('E_x =', "{0:.4g}".format(Ex),'V/m')
print('E_y =', "{0:.4g}".format(Ey),'V/m')
print('E   =', "{0:.4g}".format(E),'V/m\n')

print('(x0, y0)   = (',"{0:.4g}".format(x0),',',"{0:.4g}".format(y0),') m')
print('(x_c, y_c) = (',"{0:.4g}".format(x_c),',',"{0:.4g}".format(y_c),') m\n')

print('(vx0, vy0)   = (',"{0:.4g}".format(vx0),',',"{0:.4g}".format(vy0),') m/s')
print('|v0|         =',"{0:.4g}".format(v0),'m/s')
print('(v_x*, v_y*) = (',"{0:.4g}".format(vx_star),',',"{0:.4g}".format(vy_star),') m/s')
print('|v*|         =',"{0:.4g}".format(v_star),'m/s')
print('|V|          =',"{0:.4g}".format(V),'m/s\n')

print('(ax0, ay0) = (',"{0:.4g}".format(ax0),',',"{0:.4g}".format(ay0),') m/s^2')
print('|a0|       =',"{0:.4g}".format(a0),'m/s^2\n')

print('τ =',"{0:.4g}".format(τ),'s')
print('T =',"{0:.4g}".format(T),'s')
print('ω =',"{0:.4g}".format(ω),'rad/s')
print('θ =',"{0:.4g}".format(θ*180/np.pi),'°')
````
