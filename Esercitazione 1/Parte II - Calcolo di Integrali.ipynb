Calcolare i seguenti integrali

$\displaystyle I = \int_0^1 x^2 \;dx  \;\;\;\;\;\; I = \int_0^2 x^2 \;dx \;\;\;\;\;\; I = \int_0^1 \cos x \;dx$

con il metodo Monte Carlo della media. Eseguire per ogni integrale $K=100$ campionamenti ognuno di dimensione crescente $N = 10,\ 50,\ 250,\ 1250,\ 6250$. Riportare la stima dell'errore sul piano $\sigma - 1/\sqrt N$ e verificare che per ogni diversa funzione i punti si dispongono lungo una retta.

$\displaystyle I_k(N) = (b-a) \frac{1}{N} \sum_{i=1}^N f(x_i) \;\;\;\;\;\;\;\;\;\;
(\Delta \hat I)^2 = \frac{1}{K} \sum_{k=1}^K I_k(N)^2 - \left( \frac{1}{K} \sum_{k=1}^K I_k(N) \right)^2$

La teoria statistica prescrive che $(\Delta \hat I)(N) = \sigma/ \sqrt N$, essendo:
- $\Delta \hat I$ la dispersione (o deviazione standard) del campione di medie $\bigg\{\displaystyle \hat{I_k} = \frac{(b-a)\; \sum_{i=1}^N f(x_i)}{N} \bigg\}_{k\leqslant K}$;
- $N$ la grandezza del campione iniziale $\{(b-a)\; f(x_i) \}$;
- $\sigma$ la dispersione del campione $\{(b-a)\; f(x_i) \}_{i\leqslant N}$, dipendente dalla distribuzione di $\{ f(x_i) \}_{i\leqslant N}$.

Ma come si distribuisce $f(x_i)$? Per fare un confronto con la teoria, è necessario conoscere le funzioni densità di probabilità di $\varphi_1(x)=\varphi_2(x)=x^2$ e $\varphi_3(x)=\cos x$, con $x$ una variabile casuale distribuita uniformemente in $[0,1]$ ($[0,2]$ nel caso di $\varphi_2$).

In generale, se $\mathbb{E}\left[(x-\mu_\varphi)^2\right] \equiv \sigma_\varphi^2$ è il momento del second'ordine della distribuzione della variabile casuale $\varphi(x)$, la deviazione standard di $(b-a)\; \varphi(x)$ sarà $(b-a)\sigma_\varphi$ (essendo la deviazione standard omogenea). Ricordiamo che $(b-a)=1$ nel primo e terzo caso, mentre $(b-a)=2$ nel secondo. Il problema si riconduce quindi al calcolo di $\sigma_{\varphi}$ in ognuno dei tre casi. Facciamo allora una digressione.


Sia $x$ una variabile casuale con funzione densità di probabilità $p(x), \; x\in [a,b]$. Sia $\varphi: x \in [a,b] \rightarrow y=\varphi (x) \in \mathbb{R}$. Vogliamo determinare la funzione di probabilità $f(y)$ della variabile casuale $y=\varphi (x)$, note $p$ e $\varphi$.

Assumiamo per semplicità che $\varphi$ sia monotona crescente (è il caso di $\varphi (x)=x^2$ in $[0,1]$ e in $[0,2]$). Allora $\varphi ([a,b]) = [\varphi(a),\varphi(b)]$. Inoltre la probabilità conserva la monotonia: fissiamo $x$ e $y=\varphi(x)$. Varrà che

$P[y' \leqslant y] = P[x' \leqslant x] \;\;\;$ ove $x'\in[a,b]$ e $y'=\varphi(x') \in [\varphi(a),\varphi(b)]$ sono variabili. Sviluppiamo l'identità e riportiamo le espressioni in funzione di $y$:

$\displaystyle \int_{\varphi(a)}^y f(y')dy' = \int_a^x p(x')dx'$

Sia $\eta \equiv \varphi^{-1}$. Possiamo porre la sostituzione $x'=\eta(y')$ nel secondo integrale, da cui $dx' = \dot{\eta}(y')dy'; \;\; y'=\varphi(x'); \;\; [a,x] \rightarrow [\varphi(a),\varphi(x)]=[\varphi(a),y]$

$\displaystyle \int_{\varphi(a)}^y f(y')dy' = \int_{\varphi(a)}^y p(\eta(y')) \; \dot{\eta}(y')dy'$

Da cui, derivando ambo i membri in $y$, si ricava

$f(y) = p(\eta (y))\; \eta'(y)\;\;$ con $\;\; \eta=\varphi^{-1}$

È immediato verificare che $f(y)$ è normalizzata:

$\displaystyle \int_{\varphi(a)}^{\varphi(b)} p(\eta(y))\; \eta'(y) dy = \int_a^b p(x) dx = 1 \;\;$ con la sostituzione $x\equiv\eta(y)$, ossia $y=\varphi(x)$.

Nel caso in cui $\varphi$ sia monotona ma decrescente, si ha $\varphi([a,b])=[\varphi(b),\varphi(a)]$ e la probabilità di $\varphi(x)$ inverte la monotonia:

$P[y' \leqslant y] = P[x' \geqslant x]$ da cui

$\displaystyle \int_{\varphi(b)}^y f(y')dy' = \int_x^b p(x')dx' = -\int_b^x p(x')dx'$

per cui stavolta, ripetendo i calcoli, si ottiene

$f(y) = -p(\eta (y))\; \eta'(y)\;\;$ essendo $\eta'<0$, visto che $\varphi$ è decrescente.

Dunque, in generale: $\fbox{$ f(y) = p(\eta (y))\; |\eta'(y)| \;\;$ con $\;\; \eta=\varphi^{-1} $}$ a patto che $\varphi$ sia *derivabile* e *strettamente monotona* (quindi invertibile con inversa derivabile).

Questo risultato (insieme alle argomentazioni usate) permette anche di dimostrare il ***teorema della media***:

*Sia $x$ una variabile casuale uniforme distribuita in $[a,b]$ e sia $\varphi: x \in [a,b] \rightarrow y=\varphi(x) \in \mathbb{R}$* (assumiamo $\varphi$ monotona crescente per semplicità)*. Allora il valore atteso di $\varphi(x)$ è pari alla media integrale di $\varphi$:*

$\displaystyle \mathbb{E}[\varphi(x)]=\frac{1}{b-a}\int_a^b \varphi(x)\;dx \equiv \langle \varphi \rangle_{[a,b]}$

***Dim.:*** Sappiamo che $\varphi(x)$ si distribuisce secondo la funzione di probabilità $f(y) = p(\eta(y))\;\eta'(y)=\frac{1}{b-a}\; \eta'(y)$. Quindi:

$\displaystyle \mathbb{E}[\varphi(x)] \equiv \int_{\varphi(a)}^{\varphi(b)} y\; f(y) dy = \frac{1}{b-a} \int_{\varphi(a)}^{\varphi(b)} y \; \eta'(y) dy. \;\;$ sostituiamo: $\;\; y\equiv \varphi(x) \;\; \Rightarrow \;\; x=\eta(y), \;\; dx=\eta'(y)dy, \;\; [\varphi(a),\varphi(b)] \rightarrow [a,b]$:

$\displaystyle \mathbb{E}[\varphi(x)] \equiv \mu_{\varphi} = \frac{1}{b-a} \int_a^b \varphi(x)dx = \langle \varphi \rangle_{[a,b]} \;\;\;\;\;\;\; \square$

Per calcolare $\sigma_{\varphi}$ possiamo avvalerci di una formula semplificata:

$\displaystyle \sigma_{\varphi}^2 = \mathbb{E}\left[(\varphi(x)-\mu_\varphi)^2\right] = \int_{\varphi(a)}^{\varphi(b)}(y-\mu_{\varphi})^2\; f(y)dy = \int_{\varphi(a)}^{\varphi(b)} (y^2-2\mu_{\varphi}y + \mu_{\varphi}^2)\; f(y)dy = \\ \displaystyle \int_{\varphi(a)}^{\varphi(b)}y^2\; f(y)dy - 2\mu_{\varphi}\int_{\varphi(a)}^{\varphi(b)}y \; f(y) dy + \mu_{\varphi}^2 \int_{\varphi(a)}^{\varphi(b)}f(y)dy = \int_{\varphi(a)}^{\varphi(b)}y^2\; f(y)dy - 2\mu_{\varphi}^2+\mu_{\varphi}^2 = \int_{\varphi(a)}^{\varphi(b)} y^2\; f(y)dy - \mu_{\varphi}^2 = \mathbb{E}[y^2] - \mathbb{E}[y]^2$

ove $\; \mathbb{E}[y]^2 \equiv \mathbb{E}[\varphi(x)]^2 =\langle \varphi \rangle^2$, mentre $\mathbb{E}[y^2]$ si può ricondurre a $\varphi(x)$ con la stessa sostituzione usata prima:

$\displaystyle \mathbb{E}[y^2]=\int_{\varphi(a)}^{\varphi(b)} y^2\; f(y)dy = \int_{\varphi(a)}^{\varphi(b)}y^2 \; \frac{\eta'(y)}{b-a}dy = \Big[y \equiv \varphi(x)  \Big]= \frac{1}{b-a} \int_{a}^{b}\varphi^2(x)dx = \langle\varphi^2\rangle$. Da cui

$ \fbox{$\sigma_{\varphi}^2 = \langle\varphi^2\rangle - \langle\varphi\rangle^2$}$

Analizziamo ora in dettaglio i tre casi:

- *caso 1*:

$y=\varphi(x)=x^2 \;\;\;\; x \in [0,1] \\
\displaystyle f(y)= \frac{1}{1-0} |\eta'(y)| = \frac{d}{dy} \sqrt{y} = \frac{1}{2\sqrt{y}} \;\;\;\; y \in [\varphi(0), \varphi(1)] =[0,1] \\
\displaystyle \mu_{\varphi} = \int_0^1 y\; f(y)dy = \int_0^1 \frac{\sqrt{y}}{2}dy = \frac{1}{3}$

Notiamo che $\displaystyle \mu_{\varphi}\cdot (b-a)= \mu_{\varphi} =\frac{1}{3} = \int_0^1 x^2 dx$; compatibilmente col teorema della media.

$\displaystyle \sigma_{\varphi}^2= \langle\varphi^2\rangle - \langle\varphi\rangle^2 = \int_0^1 x^4 dx - \left(\int_0^1 x^2dx \right)^2 = \frac{1}{5}-\frac{1}{3^2}=\frac{4}{45} \; \Rightarrow \; \sigma_{\varphi}=\frac{2}{\sqrt{45}} \;\;\; \Rightarrow \;\;\; \displaystyle \sigma_1=(b-a)\; \sigma_{\varphi}=\frac{2}{\sqrt{45}} $

- *caso 2*:

$y=\varphi(x)=x^2 \;\;\;\; x \in [0,2] \\
\displaystyle f(y)= \frac{1}{2-0} |\eta'(y)| = \frac{1}{2}\frac{d}{dy} \sqrt{y} = \frac{1}{4\sqrt{y}} \;\;\;\; y \in [\varphi(0), \varphi(2)] =[0,4] \\
\displaystyle \mu_{\varphi} = \int_0^4 y\; f(y)dy = \int_0^4 \frac{\sqrt{y}}{4}dy = \frac{4}{3}$

E infatti, $\displaystyle \mu_{\varphi}\cdot (b-a)= 2\mu_{\varphi} =\frac{8}{3} = \int_0^2 x^2 dx$; in accordo col teorema.

$\displaystyle \sigma_{\varphi}^2= \langle\varphi^2\rangle - \langle\varphi\rangle^2 = \frac{1}{2}\int_0^2 x^4 dx - \left(\frac{1}{2}\int_0^2 x^2dx \right)^2 = \frac{16}{5}-\frac{16}{9}=\frac{64}{45} \; \Rightarrow \; \sigma_{\varphi}=\frac{8}{\sqrt{45}} \;\;\; \Rightarrow \;\;\; \sigma_2=(b-a)\; \sigma_{\varphi}=\frac{16}{\sqrt{45}} $

- *caso 3*:

$y=\varphi(x)=\cos x \;\;\;\; x \in [0,1] \;\;$ (funzione monotona decrescente!)

$\displaystyle f(y)= \frac{1}{1-0} |\eta'(y)| = \bigg| \frac{d}{dy} \cos^{-1}y\; \bigg| = \frac{1}{\sqrt{1-y^2}} \;\;\;\; y \in [\varphi(1), \varphi(0)] =[\cos 1,0] \\
\displaystyle \mu_{\varphi} = \int_{\cos 1}^1 y\; f(y)dy = \int_{\cos 1}^1 \frac{y}{\sqrt{1-y^2}}dy = -\left[ \sqrt{1-y^2} \right]_{\cos 1}^1 = \sin 1$

Anche in questo caso, $\displaystyle \mu_{\varphi}\cdot (b-a)= \mu_{\varphi} =\sin 1 = \int_0^1 \cos x dx$.

$\displaystyle \sigma_{\varphi}^2= \langle\varphi^2\rangle - \langle\varphi\rangle^2 = \int_0^1 \cos^2x dx - \left(\int_0^1 \cos x dx \right)^2 = \left[\frac{x}{2} + \frac{\sin 2x}{4}\right]_0^1 - \sin^2 1 = \\
\displaystyle \frac{1}{2} + \frac{\sin 2}{4} - \sin^2 1 = \frac{1}{2} + \frac{\sin 2}{4} - \frac{1-\cos 2}{2} = \frac{\sin2 + 2\cos 2}{4} \; \Rightarrow \; \sigma_{\varphi}=\frac{\sqrt{\sin 2 + 2\cos 2}}{2} \;\;\; \Rightarrow \;\;\; \sigma_3=(b-a)\; \sigma_{\varphi}=\frac{\sqrt{\sin 2 + 2\cos 2}}{2} $

In definitiva, $\displaystyle \fbox{$ \sigma_1=\frac{2}{\sqrt{45}}, \;\;\; \sigma_2=\frac{16}{\sqrt{45}}, \;\;\; \sigma_3=\frac{\sqrt{\sin 2 + 2\cos 2}}{2}$}$

$\sigma_1,\sigma_2,\sigma_3$ sono le pendenze teoriche delle rette $\displaystyle (\Delta\hat{I})_{1,2,3} \left( 1/\sqrt{N} \right)=\frac{\sigma_{1,2,3}}{\sqrt{N}}$ che graficheremo.

Con le stesse argomentazioni, si proverebbe che per $\varphi(x)=x^2, \; x\in [0,a]$ risulta $\sigma_{\varphi}^2=\frac{4}{45}a^4 \;\;$, da cui $\displaystyle \fbox{$ \sigma_a=\frac{2}{\sqrt{45}}a^3 $}$

```python
import matplotlib.pyplot as plt
import numpy as np
import random as ran
from time import time

seed0=721
Nlim=10**6
ε2=10**(-8)                             #possiamo scegliere una tolleranza diversa dal punto precedente
K=500
Dimensions = [10,50,250,1250,6250]      #lista delle dimensioni N assegnate per traccia (opzione 1)

X_N = np.arange(0.01,0.16,0.01)
Xmin = X_N.min()
Xmax = X_N.max()
Dimensions = 1/X_N**2
Dimensions = Dimensions.astype('int')   #dimensioni N in funzione di alcune ascisse scelte (quelle in X_N), in modo da
print('N =',Dimensions)                 #avere arbitrarietà sulle posizioni dei punti nel grafico σ -vs- 1/√N. In questo
                                        #modo è possibile realizzare un grafico più simmetrico e con un numero arbitrario
                                        #di 'punti sperimentali' (opzione che userò)

a=1                                     #nel caso 1 si integrerà in [0,a]; nel caso 2 in [0,b]. È una generalizzazione
b=2                                     #(opzionale) della traccia.
```

```python
ran.seed(seed0)
f = lambda x: x**2                             #caso 1
I1 = a**3/3                                    #integrale esatto
σ_1 = 2/np.sqrt(45)*a**3                       #deviazione standard teorica della VC {(b-a) f(x)}
Y_σ1=[]                                        #array delle deviazioni standard campionarie sulle medie, da graficare

t=time()
for N in Dimensions:                           #fissato N, si estraggono N VC uniformi in [a,b] per stimare l'integrale
    Sum,Sum2 = 0,0                             #(calcolando quindi la media di {(b-a) f(x_i)}). Si ripete l'operazione per
    for k in range(K):                         #K volte, realizzando K campioni e K medie. Quindi è possibile calcolare la
        S=0                                    #deviazione standard campionaria sulle K medie (ΔI). Questa sarà graficata
        for i in range(N):                     #in funzione di 1/√N, e ci si aspetterà una dipendenza lineare.
            x=a*ran.random()
            S+=f(x)
        I_k = a*S/N
        Sum+=I_k
        Sum2+=I_k**2
    ΔI = np.sqrt(Sum2/K - (Sum/K)**2)
    Y_σ1.append(ΔI)


Y1_sample=[]                                   #in questa fase si segue la stessa logica della prima parte: si vuole
Y_I1=[]                                        #istogrammare la distribuzione {f(x_i)} confrontandola con quella teorica
Yε_I1=[]                                       #(array Y1_sample), rappresentare l'andamento della stima dell'integrale al
x=a*ran.random()                               #crescere di N (array Y_I1) nonché l'errore relativo dell'integrale (array
y=f(x)                                         #Yε_I1).
n=1
S=y
I_est=a*S

Y1_sample.append(y)
Y_I1.append(I_est)
Yε_I1.append(abs(I_est - I1)/I1)

while abs(I_est - I1)/I1>=ε2:
    x=a*ran.random()
    y=f(x)
    n+=1
    S+=y
    I_est=a*S/n
    
    Y1_sample.append(y)
    Y_I1.append(I_est)
    Yε_I1.append(abs(I_est - I1)/I1)
    if n>=Nlim:
        break


X=np.linspace(0.001,a**2,500)                  #rappresentazione della distribuzione del campione {f(x_i)} in forma di
plt.figure(figsize=(12,7))                     #istogramma, confrontata con l'andamento teorico
plt.hist(Y1_sample,bins=100,density=True,color='red')
plt.plot(X, 1/(2*a*np.sqrt(X)),'b',lw=4,alpha=0.6, label='$f(y)=\dfrac{1}{2a\;\sqrt{y}}$')
plt.title('Distribuzione del campione $\{f(x_i)\}; \; i=1,\cdots,N=$'+str(n)+'\ncaso 1, intervallo [0,a] = [0,'+str(a)+']',color='red',size=20)
plt.legend(fontsize=14)
plt.show()

plt.figure(figsize=(12,7))                     #rappresentazione della stima dell'integrale al crescere di N
plt.plot([1,n],[I1,I1],'grey',ls='--',label='$I_{esatto}=\dfrac{a^3}{3}=$'+str(round(a**3/3,3)))
plt.plot(np.arange(1,n+1), Y_I1,'r')
plt.title('Evoluzione di $I(N); \; N_{max}=$'+str(n)+'\ncaso 1, intervallo [0,a] = [0,'+str(a)+']',color='r',size=20)
plt.xlabel('$N$',size=16)
plt.ylabel('$I(N)$',size=16,color='r')
plt.legend(fontsize=14)
plt.xscale('log')
plt.show()

plt.figure(figsize=(12,7))                     #rappresentazione della stima dell'errore al crescere di N
plt.title("Evoluzione dell'errore $\epsilon_{I}(N)$\ncaso 1, intervallo [0,a] = [0,"+str(a)+"]",color='darkred',size=20)
plt.plot(np.arange(1,n+1),Yε_I1,'darkred',lw=1,label='$\epsilon_{I}(N)$')
plt.xlabel('$N$',size=16)
plt.ylabel('$\epsilon_{I}(N)$',size=16)
plt.legend(fontsize=14)
plt.xscale('log')
plt.yscale('log')
plt.show()

plt.figure(figsize=(12,7))                     #rappresentazione di ΔI in funzione di 1/√N, confrontata con la retta ideale
plt.plot([Xmin,Xmax],[Xmin*σ_1,Xmax*σ_1],'grey',ls='--',
         label='$(\Delta\hat{I})\;(1/\sqrt{N}) \;=\; \dfrac{\sigma_1}{\sqrt{N}} \;=\; \dfrac{2\; a^3}{\sqrt{45}} \; \dfrac{1}{\sqrt{N}}$')
plt.plot(X_N,Y_σ1,'o r')
plt.title('Andamento  $\Delta\; \hat{I}$  -vs-  $1/\sqrt{N}$\ncaso 1, intervallo [0,a] = [0,'+str(a)+']',color='r',size=20)
plt.xlabel('$1/\sqrt{N}$',size=16)
plt.ylabel('$\Delta\; \hat{I}$',size=16,color='r')
plt.legend(fontsize=14)
plt.show()

t=time()-t

if n>=Nlim:
    print('Grandezza del campione: N >',n,'(!)')
else: print('Grandezza del campione: N =',n)
print('Tolleranza relativa:    ε =',ε2)
print('Tempo di esecuzione:   Δt =',round(t,2),'s')

print('\nIntegrale stimato:     I_est =',round(I_est,10),'; esatto:  I =',round(I1,10))
print('Errore relativo su I:   ε(I) =',round((I_est-I1)/I1,10))
```
