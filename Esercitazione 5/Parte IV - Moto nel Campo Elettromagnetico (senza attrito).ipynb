### Parte IV: Moto nel Campo Magnetico e Elettrico (senza attrito)

Studiare il moto di un elettrone in un campo magnetico ortogonale al vettore velocità iniziale. L'equazione del moto è

$\Large \displaystyle m \ \frac{d \vec{v}}{dt} = q \; \left(\vec{v} \times \vec{B} + \vec{E} \right)$

Calcolare numericamente e rappresentare la traiettoria dell'elettrone con un campo magnetico costante, ortogonale alla velocità e il cui modulo iniziale è dato. Partire dal caso $\vec{E} = \textbf{0}$; successivamente variare $\vec{B}$. Studiare poi gli stessi casi aggiungendo un campo elettrico uniforme.

$ \displaystyle \vec{B} \equiv (0, \; 0, \; B), \;\;\;\; \vec{E} \equiv (E_x, \; E_y, \; 0)$

$\displaystyle \vec{r}(0) \equiv (x_0, \; y_0, \; 0), \;\;\;\; \vec{v}(0) \equiv \left( v_x^0, \; v_y^0, \; 0 \right)$

Le equazioni del moto in forma normale sono

$\begin{cases} \dot{x} = v_x \\ \\ \dot{y} = v_y \\ \\ \dot{v}_x = \omega \ v_y + q E_x/m \\ \\ \dot{v}_y = -\omega \ v_x + q E_y/m \end{cases} \;\;\;\;$ essendo $\displaystyle \omega \equiv \frac{qB}{m}$ ($\omega$ ha un segno che dipende dalla carica $q$ e dalla proiezione lungo l'asse $z$ di $\vec{B}$).

Il sistema può essere inizialmente risolto in $v_x(t), v_y(t)$ disaccoppiando le eq. $3,4$: derivando la $3$ e sostituendo a $\dot{v}_y$ l'espressione della $4$, si ottiene un'equazione lineare in $v_x$; si procede analogamente per $v_y$. Imponendo le dovute condizioni iniziali si ricava

$\displaystyle v_x(t) = \;\;\;\frac{E_y}{B} + \left( v_x^0 - \frac{E_y}{B} \right) \, \cos \omega t + \left( v_y^0 + \frac{E_x}{B} \right) \, \sin \omega t \\
 \displaystyle v_y(t) = -\frac{E_x}{B} + \left( v_y^0 + \frac{E_x}{B} \right) \, \cos \omega t - \left( v_x^0 - \frac{E_y}{B} \right) \, \sin \omega t$

L'espressione può essere resa più compatta con il formalismo matriciale: posto

$\displaystyle \vec{v}(t) \equiv (v_x(t), \; v_y(t))^t$

$\displaystyle \vec{v}_c \equiv \left( \frac{E_y}{B}, -\frac{E_x}{B} \right)^t$

$\displaystyle \vec{v}_0 \equiv \left(v_x^0, \; v_y^0 \right)^t$

e sia $\displaystyle \mathcal{R}_{\omega}(t) \equiv \begin{pmatrix} \cos \omega t & \sin \omega t \\ -\sin \omega t & \cos \omega t \end{pmatrix}$ la matrice delle rotazioni, è immediato verificare che vale $ \vec{v}(t) = \vec{v}_c + \mathcal{R}_{\omega}(t) \cdot \left(\vec{v}_0 - \vec{v}_c \right)$

Geometricamente, vuol dire che il vettore velocità ruota con pulsazione $\omega$ intorno al centro $\vec{v}_c$, il quale si riduce all'origine per $\vec{E} = \textbf{0}$. $\vec{v}_c$ è un termine di "offset" della velocità: $v_x$ e $v_y$ oscillano periodicamente ma non attorno il valore di equilibrio $\textbf{0}$ (a meno che il campo elettrico non sia nullo); possiamo quindi aspettarci che  il moto della particella carica si componga di due contributi: un moto rettilineo uniforme lungo $\vec{v}_c$ (che notiamo essere ortogonale ad $\vec{E}$) e un moto circolare uniforme. La traiettoria sarà una spirale "schiacciata" sul piano $x,y$.

Derivando $v_x$ e $v_y$, ricaviamo anche l'accelerazione iniziale:

$\displaystyle a_x^0 = \omega \left(v_y^0 + \frac{E_x}{B} \right), \;\;\; a_y^0 = -\omega \left(v_x^0 - \frac{E_y}{B}\right)$

Con opportuni magheggi algebrici, è possibile anche esprimere le velocità come

$\begin{cases} \displaystyle v_x(t) = \frac{E_y}{B} + V \sin(\omega t + \phi) \\ \\ \displaystyle v_y(t) = -\frac{E_x}{B} + V \cos(\omega t + \phi) \end{cases} \;\;$

avendo posto $\displaystyle V \equiv \sqrt{ \left( v_x^0 - \frac{E_y}{B} \right)^2 + \left( v_y^0 + \frac{E_x}{B} \right)^2 }$ e $\displaystyle \phi \equiv \arctan_2 \left( v_x^0 - \frac{E_y}{B}, \; v_y^0 + \frac{E_x}{B} \right)$ ($\arctan_2$ è l'[arcotangente a due argomenti](https://it.wikipedia.org/wiki/Arcotangente2)). Da cui, per derivazione e integrazione, si ricava

$\begin{cases} \displaystyle a_x(t) = \omega V \cos(\omega t + \phi) \\ \\ \displaystyle a_y(t) = -\omega V \sin(\omega t + \phi) \end{cases}$

$\begin{cases} \displaystyle x(t) = \left( x_0 + \frac{v_y^0 + E_x/B}{\omega} \right) + \frac{E_y}{B}t - \frac{V}{\omega} \cos(\omega t + \phi) \\ \\
\displaystyle y(t) = \left( y_0 - \frac{v_x^0 - E_y/B}{\omega} \right) - \frac{E_x}{B}t + \frac{V}{\omega} \sin(\omega t + \phi) \end{cases}$

Nel caso semplice $\vec{E} = \mathbf{0}$ vale $V = \sqrt{\left( v_x^0 \right)^2 + \left( v_y^0 \right)^2} \equiv v_0, \;\;\; \phi = \arctan_2(v_x^0, \; v_y^0)$, da cui

$\begin{cases} \displaystyle x(t) = \left(x_0 + \frac{v_y^0}{\omega} \right) - \frac{v_0}{\omega} \cos(\omega t + \phi) \\ \\ \displaystyle y(t) = \left(y_0 - \frac{v_x^0}{\omega} \right) + \frac{v_0}{\omega} \sin(\omega t + \phi) \end{cases}$

Posto $\displaystyle \vec{r}_c \equiv (x_c, y_c) \equiv \left( x_0 + \frac{v_y^0}{\omega}, y_0 - \frac{v_x^0}{\omega} \right)$, è immediato verificare che $\displaystyle |\vec{r}(t) - \vec{r}_c| = \frac{v_0}{|\omega|} = \frac{mv_0}{|qB|} \equiv R \;\;\; \forall t \geqslant 0$. Ossia, la traiettoria è una circonferenza centrata in $\displaystyle \vec{r}_c$; $R$ è il ***raggio di Larmor***, mentre il periodo è $\displaystyle T = \frac{2\pi}{|\omega|} = 2\pi \frac{m}{|qB|}$.

Le velocità sono

$\begin{cases} \displaystyle v_x(t) = v_0 \sin(\omega t + \phi) \\ \\ \displaystyle v_y(t) = v_0 \cos(\omega t + \phi)\end{cases}$

e le accelerazioni

$\begin{cases} \displaystyle a_x(t) = \omega v_0 \cos(\omega t + \phi) \\ \\ \displaystyle a_y(t) = -\omega v_0 \sin(\omega t + \phi)\end{cases}$

L'accelerazione iniziale si riduce a

$a_x^0 = \omega v_y^0, \;\;\; a_y^0 = -\omega v_x^0$

Notiamo che a prescindere dalla presenza del campo elettrico, il modulo di $\vec{a}(t)$ rimane costante:

$\displaystyle |\vec{a}(t)| \equiv a = |\omega| V \;\; ( = |\omega| v_0$ nel caso $\vec{E} = \mathbf{0})$

### Considerazioni energetiche

Sulla particella agisce la forza elettrica e la forza di Lorentz: la prima, almeno nel caso statico, ammette un potenziale; la seconda non compie lavoro. Ci aspettiamo quindi che l'energia meccanica del sistema si conservi, e potremmo verificarlo numericamente:

$\displaystyle \mathcal{E}(t) \equiv K(t) + U(t)$

ove

$\displaystyle K(t) = \frac{1}{2} m v^2(t) = \frac{1}{2} m \left( v_x^2(t) + v_y^2(t) \right)$

$\displaystyle U(t) \equiv U(\vec{r}(t))$

Sia $\vec{r}_z$ la posizione di riferimento del potenziale (arbitraria), allora varrà

$\displaystyle U(\vec{r}) - U(\vec{r}_z) = -q \int_{\vec{r}_z}^{\vec{r}} \vec{E} \cdot d\vec{r} = -q \ \vec{E} \cdot \int_{\vec{r}_z}^{\vec{r}} d\vec{r} = -q \ \vec{E} \cdot \left( \vec{r} - \vec{r}_z  \right)$

Per comodità scegliamo $\vec{r}_z \equiv \vec{r}_0 = \left(x_0, y_0 \right)$ e $U(\vec{r}_z) \equiv 0$. Da cui

$\displaystyle U(\vec{r}) = -q \left[ E_x (x - x_0) + E_y (y - y_0)\right]$. Osserviamo che il potenziale è costante a patto che $E_x (x - x_0) + E_y (y - y_0) \equiv V_0$: si tratta dell'equazione di una retta ortogonale al vettore $(E_x, E_y) = \vec{E}$. Ossia, compatibilmente con la teoria, le linee equipotenziali del sistema sono ortogonali al campo. In definitiva:

$\displaystyle \mathcal{E}(t) = \frac{1}{2}m \left( v_x^2(t) + v_y^2(t) \right) - q \left[ E_x (x(t) - x_0) + E_y (y(t) - y_0)\right]$

Con i calcoli non è difficile provare che $\displaystyle \mathcal{E}(t) \equiv \mathcal{E}_0 = \frac{1}{2}mv_0^2$ (le energie cinetica e potenziale oscillano in opposizione di fase attorno due valori costanti che, se ricombinati, restituiscono proprio $K_0$). Noi lo faremo usando l'algoritmo.

In presenza di un campo $\vec{E}$, può sembrare paradossale che una particella carica, pur con moti oscillatori, nel complesso trasli ortogonalmente ad $\vec{E}$. In realtà una traiettoria simile è compatibile sia con la fisica che c'è dietro, sia con la conservazione dell'energia: il campo elettrico accelera la particella nella direzione di $\vec{E}$; l'aumento del modulo della velocità porta all'aumento della forza di Lorentz, che però è ortogonale alla velocità; quindi questa devia immediatamente la particella dalla direzione dei $E$. Gli effetti dei due campi competono quindi l'uno con l'altro. Allo stesso tempo, dato che $|\vec{v}|$ non aumenta indefinitamente in modulo, è naturale ritrovare che $K(t)$ oscilli attorno un offset; stessa cosa per $U(t)$. Sommando i due termini, otteniamo la quantità costante $\mathcal{E}_0$. Svolgendo i conti, e osservando che vale

- $\displaystyle V^2 = \left( v_x^0 - \frac{E_y}{B} \right)^2 + \left( v_y^0 + \frac{E_x}{B} \right)^2 = v_0^2 + \frac{E^2}{B^2} + 2 \, \frac{E_x v_y^0 - E_y v_x^0}{B}$
- $\displaystyle \frac{E^2}{B^2} = v_c^2$

otteniamo

$\displaystyle K(t) = \frac{1}{2}m \ \left[ \left( \frac{E^2}{B^2} + V^2 \right) + 2 \frac{V}{B} \; \left( E_y \sin(\omega t + \phi) - E_x \cos(\omega t + \phi) \right) \right] \\
 \displaystyle U(t) = -m \ \left[ \left( \frac{E^2}{B^2} + \frac{E_x v_y^0 - E_y v_x^0}{B} \right) + \frac{V}{B} \; \left( E_y \sin(\omega t + \phi) - E_x \cos(\omega t + \phi) \right) \right]$

I contributi oscillanti si compensano, mentre i termini di offset sono

$\displaystyle K_{ofs} = \frac{1}{2}m \ \left[ \frac{E^2}{B^2} + V^2 \right] = \frac{1}{2} m \ \left(v_c^2 + V^2 \right) \\
 \displaystyle U_{ofs} = -m \ \left[ \frac{E^2}{B^2} + \frac{E_x v_y^0 - E_y v_x^0}{B} \right] = -\frac{1}{2}m \ \left[ 2 \, \frac{E^2}{B^2} + 2 \, \frac{E_x v_y^0 - E_y v_x^0}{B} \right] = \frac{1}{2}m \ \left( v_0^2 - v_c^2 - V^2 \right)$

Da cui

$\displaystyle K_{ofs} + U_{ofs} = \frac{1}{2} mv_0^2 = \mathcal{E}_0$
